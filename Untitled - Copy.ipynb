```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

def get_data(start, end, stats,os,lst_country,top1,top5, unit):

    countries = stats['Country'].unique()
    stats = stats[(stats['Date'] > start) & (stats['Date'] < end)]
    for country in countries:
        days1 = stats[(stats['Country'] == country) & (stats['Overall'] == 1)]['Overall'].count()
        days5 = stats[(stats['Country'] == country) & (stats['Overall'] <5)]['Overall'].count()
        lst_country.append(country)
        top1.append(days1)
        top5.append(days5)
        unit.append(os)
    
    return  lst_country, top1, top5, unit
    
#read all the sheets
xl = pd.ExcelFile('Ragnarok ranking.xlsx')
hk_iphone = xl.parse('HK iPhone')
hk_android = xl.parse('HK android')
sea_android = xl.parse('SEA android')
sea_iphone = xl.parse('SEA iPhone')
rest_iphone = xl.parse('Amerika&Oceana iPhone')
rest_android = xl.parse('Amerika&Oceana android')

#replace the NaT  with date in the Date-column
sea_android = sea_android.fillna(method = 'ffill')
sea_iphone = sea_iphone.fillna(method = 'ffill')
hk_iphone = hk_iphone.fillna(method = 'ffill')
hk_android = hk_android.fillna(method = 'ffill')
rest_iphone = rest_iphone.fillna(method = 'ffill')
rest_android = rest_android.fillna(method = 'ffill')

#drop Unnamed columns
hk_iphone = hk_iphone.drop(columns = ['Unnamed: 3','Unnamed: 5','Unnamed: 7','Unnamed: 9'])
hk_android = hk_android.drop(['Unnamed: 3','Unnamed: 5','Unnamed: 7','Unnamed: 9','Role Playing (Games)'], axis = 1)
sea_iphone = sea_iphone.drop(['Unnamed: 3','Unnamed: 5','Unnamed: 7'], axis = 1)
sea_android = sea_android.drop(['Unnamed: 3','Unnamed: 5','Unnamed: 7'], axis = 1)
rest_iphone = rest_iphone.drop(['Unnamed: 3','Unnamed: 5','Unnamed: 7','Unnamed: 9'], axis = 1)
rest_android = rest_android.drop(['Unnamed: 3','Unnamed: 5','Unnamed: 7'], axis = 1)

q4_1 = '2018-11-05'
q4_2 = '2018-12-31'

q1_1 = '2019-01-01'
q1_2 = '2019-03-31'

#Q4
start_q4 = q4_1
end_q4 = q4_2
lst_country_q4 = []
top1_q4 = []
top5_q4 = []
os_q4 = []
#Q1
start_q1 = q1_1
end_q1 = q1_2
lst_country_q1 = []
top1_q1 = []
top5_q1 = []
os_q1 = []

#Q1 western market
start_q1_west = q1_1
end_q1_west = q1_2
lst_country_q1_west = []
top1_q1_west = []
top5_q1_west = []
os_q1_west = []

#append countries, ranking och OS for Q4
lst_country_q4, top1_q4, top5_q4, os_q4 = get_data(start_q4,end_q4,hk_iphone,'iphone',lst_country_q4, top1_q4, top5_q4, os_q4)
lst_country_q4, top1_q4, top5_q4, os_q4 = get_data(start_q4,end_q4,hk_android,'android',lst_country_q4, top1_q4, top5_q4, os_q4)
lst_country_q4, top1_q4, top5_q4, os_q4 = get_data(start_q4,end_q4,sea_iphone,'iphone',lst_country_q4, top1_q4, top5_q4, os_q4)
lst_country_q4, top1_q4, top5_q4, os_q4 = get_data(start_q4,end_q4,sea_android,'android',lst_country_q4, top1_q4, top5_q4, os_q4)

#append countries, ranking och OS for Q1
lst_country_q1, top1_q1, top5_q1, os_q1 = get_data(start_q1,end_q1,hk_iphone,'iphone',lst_country_q1, top1_q1, top5_q1, os_q1)
lst_country_q1, top1_q1, top5_q1, os_q1 = get_data(start_q1,end_q1,hk_android,'android',lst_country_q1, top1_q1, top5_q1, os_q1)
lst_country_q1, top1_q1, top5_q1, os_q1 = get_data(start_q1,end_q1,sea_iphone,'iphone',lst_country_q1, top1_q1, top5_q1, os_q1)
lst_country_q1, top1_q1, top5_q1, os_q1 = get_data(start_q1,end_q1,sea_android,'android',lst_country_q1, top1_q1, top5_q1, os_q1)
lst_country_q1_west, top1_q1_west, top5_q1_west, os_q1_west = get_data(start_q1,end_q1,rest_iphone,'iphone',lst_country_q1_west, top1_q1_west, top5_q1_west, os_q1_west)
lst_country_q1_west, top1_q1_west, top5_q1_west, os_q1_west = get_data(start_q1,end_q1,rest_android,'android',lst_country_q1_west, top1_q1_west, top5_q1_west, os_q1_west)
```

```python
dic_q4 = {'country': lst_country_q4,
       'top1': top1_q4,
       'top5': top5_q4,
       'os': os_q4}
df_q4 = pd.DataFrame(dic_q4, columns = ['country',  'top1', 'top5','os',])

dic_q1 = {'country': lst_country_q1,
       'top1': top1_q1,
       'top5': top5_q1,
       'os': os_q1}
df_q1 = pd.DataFrame(dic_q1, columns = ['country',  'top1', 'top5','os',])

dic_q1_west = {'country': lst_country_q1_west,
       'top1': top1_q1_west,
       'top5': top5_q1_west,
       'os': os_q1_west}
df_q1_west = pd.DataFrame(dic_q1_west, columns = ['country',  'top1', 'top5','os',])
```

```python
def barplot(title, df_q4, df_q1, country):
    # width of the bars
    barWidth = 0.3

    # Choose the height of the blue bars
    bars1 = df_q4

    # Choose the height of the cyan bars
    bars2 = df_q1

    # The x position of bars
    r1 = np.arange(len(bars1))
    r2 = [x + barWidth for x in r1]
    plt.figure(figsize=(8,6))

    # Create blue bars
    plt.bar(r1, bars1, width = barWidth, facecolor='#ff9999', edgecolor = 'black', capsize=7, label='q4 (only 2m)')

    # Create cyan bars
    plt.bar(r2, bars2, width = barWidth, facecolor='#9999ff', edgecolor = 'black', capsize=7, label='q1')
    
    # Text on the bars
    for x,y in zip(r2,bars2):
         if y == 0:
            pass;
         else:
            plt.text(x, y+0.05, y, ha='center', va= 'bottom')
    for x,y in zip(r1,bars1):
        if y == 0:
            pass;
        else:
            plt.text(x, y+0.05, y, ha='center', va= 'bottom')

    plt.xticks([r + barWidth for r in range(len(bars1))], country)
    plt.ylabel('days', color = 'white')
    plt.legend()
    plt.xticks(color='white', rotation=45)
    plt.yticks(color='white')
    plt.title(title,color = 'white')
    plt.show()

def barplot_west(title, df_q1, country):
    barWidth = 0.3

    bars2 = df_q1

    # The x position of bars
    r1 = np.arange(len(bars2))

    plt.figure(figsize=(15,6))

    plt.bar(country, bars2, width = barWidth, facecolor='#9999ff', edgecolor = 'black', capsize=7, label='q1')
    
    # Text on the bars
    for x,y in zip(country,bars2):
         if y == 0:
            pass;
         else:
            plt.text(x, y+0.01, y, ha='center', va= 'bottom')

    plt.xticks([r + barWidth for r in range(len(country))], country)
    plt.ylabel('days', color = 'white')
    plt.legend()
    plt.xticks(color='white', rotation=45)
    plt.yticks(color='white')
    plt.title(title,color = 'white')
    plt.show()
    
#iPhone ranking
df_q1_iphone = df_q1[df_q1['os'] == 'iphone']
df_q4_iphone = df_q4[df_q1['os'] == 'iphone']
df_q1_west_iphone = df_q1_west[df_q1_west['os'] == 'iphone']

#Android ranking 
df_q1_android = df_q1[df_q1['os'] == 'android']
df_q4_android = df_q4[df_q4['os'] == 'android']
df_q1_west_android = df_q1_west[df_q1_west['os'] == 'android']

barplot('Number of days Top1 iPhone',  df_q4_iphone.top1, df_q1_iphone.top1, df_q1_iphone.country)
barplot('Number of days Top5 iPhone', df_q4_iphone.top5, df_q1_iphone.top5, df_q4_iphone.country)
barplot('Number of days Top1 Android', df_q4_android.top1, df_q1_android.top1, df_q1_android.country)
barplot('Number of days Top5 Android', df_q4_android.top5, df_q1_android.top5, df_q4_android.country)
barplot_west('Number of days Top1 Android', df_q1_west_android.top1, df_q1_west_android.country)
barplot_west('Number of days Top5 Android', df_q1_west_android.top5, df_q1_west_android.country)
barplot_west('Number of days Top1 iPhone', df_q1_west_iphone.top1, df_q1_west_iphone.country)
barplot_west('Number of days Top5 iPhone', df_q1_west_iphone.top5, df_q1_west_iphone.country)
```
